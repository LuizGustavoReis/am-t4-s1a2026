## Carregar digest de base_credit

```python
import pickle
```

```python
with open('/content/drive/MyDrive/am-t4/credit.pkl', 'rb') as f:
 X_credit_treinameto , Y_credit_treinamento, X_credit_teste, Y_credit_teste = pickle.load(f)
```

```python
X_credit_treinameto.shape, Y_credit_treinamento.shape
```

```python
X_credit_teste.shape, Y_credit_teste.shape
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
naive_credit_data = GaussianNB()
naive_credit_data.fit(X_credit_treinameto, Y_credit_treinamento)
```

```python
census_predict = naive_credit_data.predict(X_credit_teste)
```

```python
census_predict
```

```python
Y_credit_teste
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
accuracy_score(Y_credit_teste, census_predict)
```

```python
confusion_matrix(Y_credit_teste, census_predict)
```

```python
from yellowbrick.classifier import ConfusionMatrix
```

```python
cm_census = ConfusionMatrix(naive_credit_data)
cm_census.fit(X_credit_treinameto, Y_credit_treinamento)

cm_census.score(X_credit_teste, Y_credit_teste)
```
